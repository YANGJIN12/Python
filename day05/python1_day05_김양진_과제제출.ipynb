1. 구구단을 2단부터 9단까지 출력하되, 계산값이 홀수인 경우만 출력하기(이중 for문)
예) 2 * 2 = 4에서 4는 짝수이므로 출력안됨
	3 * 1 = 3에서 3는 홀수이므로 출력됨

```python
for i in range(2,10):
    for j in range(1,10):
        if i * j % 2 == 1:
            print(f"{i}*{j}={i*j}")
```

4. 국어, 영어, 수학 점수를 입력받아 최고점과 최저점을 구하고 최고점와 최저 점의 차이를 출력하는 프로그램 만들기

```python
korean = english = math = 0
korean = int(input("국어 점수: "))
english = int(input("영어 점수: "))
math = int(input("수학 점수: "))

max = min = 0
if korean >= english and korean >= math:
    max = korean
elif english >= korean and english >= math:
    max = english
else:
    max = math

if korean <= english and korean <= math:
    min = korean
elif english <= korean and english <= math:
    min = english
else:
    min = math

print("최고점:", max)
print("최저점:", min)
print("최고점와 최저점의 차이:",max - min)

kor = int(input("국어 점수: "))
eng = int(input("영어 점수: "))
mat = int(input("수학 점수: "))
result = [kor, eng, mat]
result_max = int(max(result))
result_min = int(min(result))

if kor != eng != mat:
    print(f'최고점 {result_max}점, 최저점 {result_min} 두 점수의 차이는 {result_max - result_min}')
else:
    print(f'최고점 {result_max}점, 최저점 {result_min} 두 점수의 차이는 {result_max - result_min}')
```

5. 1부터 1001까지 홀수만 모두 더한 값을 출력하기(while문)

```python
sum = 0
while True:
    for i in range(1,1002,2):
        sum += i
    print("1부터 1001까지 홀수만 모두 더한 값: ", sum)
    break
    
total, num = 0, 1
while num <= 1001:
    if num % 2 == 1:
        total += num
    num += 1
print(total)
```
