2. 음료수 자판기 프로그램 만들기
1번은 '콜라', 2번은 '사이다', 3번은 '환타'이고 각 버튼에 따라 음료수 이름을 출력하고, 그 외 번호를 입력시에는 '제공하지 않는 메뉴'를 출력하기
(1, 2, 3으로 번호를 입력해도 되고, 콜라/사이다/환타로 입력해도 가능함)

```python
choice = input("음료수 선택: 1. 콜라 2. 사이다 3.환타 ")
if choice == '1' or choice == "콜라":
    print("시원한 콜라가 나왔습니다.")
elif choice == '2' or choice == "사이다":
    print("시원한 사이다가 나왔습니다.")
elif choice == '3' or choice == "환타":
    print("시원한 환타가 나왔습니다.")
else:
    print("제공하지 않는 메뉴입니다!! 다시 선택해주세요~")

#방법2
# bot = {'1':'콜라','2':'사이다','3':'환타'}
# if ((choice == '1') or (choice == '콜라')):
#     print("%s가 나옵니다."%(bot['1']))
```

3. 나이를 입력받아 입장료를 계산하는 프로그램 만들기
(단, 10세 이하의 입장료는 1000원, 65세 이상의 입장료는 0원, 기본 입장료는 2000원이다.)

```python
age = int(input("나이 입력: "))
if age <= 10:
    print("10세 이하의 입장료는 1000원입니다.")
elif age >= 65:
    print("65세 이상의 입장료는 무료입니다.")
else:
    print("기본 입장료는 2000원입니다.")
```

4. 짱구가 놀이공원에 롤러코스터를 타려고한다. 키가 120cm 미만인 사람은 롤러코스터를 타지 못하고, 키가 120~140cm인 사람은 보호자와 함께 탈 수 있고, 140cm 미만인 사람은 탈 수 있다. 짱구의 키가 129일 때 결과 값을 출력하라.

```python
height = int(input("키 입력: "))
if height < 120:
    print("탑승이 불가능합니다.")
elif height >= 120 and height <= 140:
    print("보호자와 함께 탑승하실 수 있습니다.")
else:
    print("탑승이 가능합니다.")
    
```

5. 성적 산출 프로그램을 만들고 각 결과를 출력하라.
<조건>
90점 ~ 100점 : A
80점 ~ 89점 : B
70점 ~ 79점 : C
69점 ~ 0점 : F

```python
grade = int(input("성적 입력:"))
if 90 < grade <= 100:
    print("A")
elif 90 < grade <= 89:
    print("B")
elif 70 < grade <= 79:
    print("C")
elif 0 <= grade < 69:
    print("F")
else:
    print("잘못 입력하셨습니다.")
```
