1. 리스트에 월별일수 저장해두고 1월부터 12월까지 월별 일수를 출력하는 프로그램
(zip() 튜플 형태로 출력 => 빈리스트를 하나더 만들어서 작성)
(enumerate() => 1월은 31일입니다. 2월은 28일 입니다.)

month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]


[출력결과]
1월 31일입니다.
2월 28일입니다.
3월 31일입니다.
4월 30일입니다.
5월 31일입니다.
6월 30일입니다.
7월 31일입니다.
8월 31일입니다.
9월 30일입니다.
10월 31일입니다.
11월 30일입니다.
12월 31일입니다.

```python
month = [31,28,31,30,31,30,31,31,30,31,30,31]
day = []
for i in range(1,13):
    day.append(i)
for j in zip(day, month):
    print(j)

for idx, item in enumerate(month):
    print(f'{idx+1}월은 {item}일 입니다.')
```

2. 사용자로부터 과일 이름을 입력받아 리스트에 보관하는 프로그램
(중복된 과일은 저장 불가, 중복된 과일 이름을 입력 받으면 다른과일을 입력받도록 할 것, 입력받을 수 있는 횟수는 3번)

[출력결과 예시]
어떤과일을 저장할까? : 사과
입력이 2번 남았습니다.
어떤과일을 저장할까? : 포도
입력이 1번 남았습니다.
어떤과일을 저장할까? : 포도
동일한 과일있어요.
어떤과일을 저장할까? : 복숭아
입력이 0번 남았습니다.
과일은 ['사과', '포도', '복숭아']입니다.

```python
fruits = []
count = 3
while count > 0:
    save = input("어떤과일을 저장할까?: ")
    if save in fruits:
        print("동일한 과일이 있습니다.")
        continue
    fruits.append(save)
    count -= 1 
    print(f"입력이 {count}번 남았습니다.")
print(f"과일은 {tuple(fruits)} 입니다.")
```

4. 학생 수와 학생별 파이썬, 자바, C언어 점수를 입력받고, 학생별 평균 점수를 구하는 프로그램

조건1. 평균 점수는 소수점 셋째 자리에서 반올림하여 계산
조건2. 학생의 이름과 평균점수는 딕셔너리 형태로 생성하고, 학생수는 입력받아 작성할 수 있도록 할 것
조건3. 점수는 0~100사이의 양의 정수만 입력받도록 조건 설정할 것
조건4. 0~100 사이의 양의 정수가 아니라면 '잘못된 입력입니다' 출력

[결과출력 예시(잘못된 입력없을경우)]
학생 수 입력 : 2
1번째 학생 이름 입력 : 김영선
1번째 학생의 파이썬 점수 입력 : 30
1번째 학생의 자바 점수 입력 : 40
1번째 학생의 c언어 점수를 입력 : 50
2번째 학생 이름 입력 : kys
2번째 학생의 파이썬 점수 입력 : 100
2번째 학생의 자바 점수 입력 : 60
2번째 학생의 c언어 점수를 입력 : 80
{'김영선': 40.0, 'kys': 80.0}


[결과출력 예시(잘못된 입력있을경우)]
학생 수 입력 : 1
1번째 학생 이름 입력 : 김영선
1번째 학생의 파이썬 점수 입력 : 1000
잘못된 입력입니다
1번째 학생의 파이썬 점수 입력 : -5
잘못된 입력입니다
1번째 학생의 파이썬 점수 입력 : 99
1번째 학생의 자바 점수 입력 : 400
잘못된 입력입니다
1번째 학생의 자바 점수 입력 : 50
1번째 학생의 c언어 점수를 입력 : -100
잘못된 입력입니다
1번째 학생의 c언어 점수를 입력 : 0
{'김영선': 49.67}

```python
student_dict = {}
num = int(input("학생 수 입력: "))
for count in range(num):
    score = 0
    step = len(student_dict) + 1
    name = input(f"{step}번째 학생 이름 입력: ")
    while True:
        p = int(input(f"{step}번째 학생의 파이썬 점수 입력:"))
        j = int(input(f"{step}번째 학생의 자바 점수 입력:"))
        c = int(input(f"{step}번째 학생의 C언어 점수 입력:"))
        if p >= 0 and p <= 100:
            score += p
            break
        elif j >= 0 and j <= 100:
            score += j
            break
        elif c >= 0 and c <= 100:
            score += c
            break
        else:
            print("잘못된 입력입니다.")
        student_dict[name] = round(float(score/3), 2)
print(student_dict)




       
```
