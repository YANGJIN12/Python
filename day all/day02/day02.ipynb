#1. 기본자료형

1) bool : 논리자료형(값이 있을 때 True, 값이 없을 때 False)
0,'',[],(),{} > False
False 외의 값 > True

```python
# bool

print(bool(0), type(bool(0)))
print(bool(1), type(bool(1)))
print(bool('0'), type(bool('0')))
print(bool(''),  type(bool('')))
print(10000, type(10000))
print(bool(10000), type(bool(10000)))
print(bool([]), type(bool([])))
print(bool([1]), type([1]))
```

2) str : 문자열 자료형(문자를 추출하고자 할 때 사용함)

> 인덱싱 : 문자열을 구성하는 모든 문자에 부여한 고유번호
            인덱스 시작번호는 0, 뒤에서부터의 시작번호는 -1 

> 슬라이싱 : 문자열의 인덱스를 활용해서 한 문자 이상으로 구성된 단어나 문자 추출
                슬라이싱
                string = '안녕하세요'
                string[start:stop+1;step]

```python
string = 'hello python'
print(string, type(string))  #문자열은 시퀀스 자료형이기때문에 인덱스 개념이 가능함

print(string[0])
print(string[6])
print(string[11])
print(string[-1])

print(string[0:3:1])
print(string[0:9]) #step값은 생략이 가능함 / 기본값이 1이 됨
print(string[:9]) #시작값도 생략이 가능함 / 기본값이 0이 됨[:종료값]
print(string[6:]) #맨 뒤까지 출력하고 싶으면 :으로 종료값자리를 비워줘도됨
print(string[-6:])
print(string[0::2])
print(string[-1::-1]) #증감값에 -1을 사용하게되면 역순으로 출력됨


#string2 = 'Hello World'
string2 = 'Hello World'
#슬라이싱을 이용해서 출력하기
#1. Hello
print(string2[0:6])
#2. Hello World
print(string2[:12])
#3. Hlowrd
print(string2[::2])
```

#3. 이스케이프 문자

>이스케이프 문자를 사용하는 경우
1. 문자열 사이에 '',""들어가는 경우
2. 출력할 때 줄바꿈이 아닌 tab키나 띄어쓰기로 사용하는 경우

```python
print("'아버지'가 방에 들어가신다.")
print("\"아버지\"가 방에 들어가신다.")
print("동해물과 백두산이 마르고 닳도록")
print('"동해물"과 "백두산"이 마르고 닳도록')
print("\"동해물\"과 \"백두산\"이 마르고 닳도록")
print("\t동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라만세")
print("맨 마지막에는 end옵션이 들어간다." + "end옵션내에 이스케이프문자를 사용할 수 있습니다.", end='\t')
print("마지막 수업입니다.")
```

#4. 형식을 갖춘 문자열

>%연산자

>.format()

>f_string

```python
#%연산자
print('%d' % 10, type(10))
print('%f' % 10, type(float(10)))
print('%s' % 'python', type('python'))

number10 = 10
number11 = 20
print('%d %d' %(number10, number11))

#.format()
print('{} + {} = {}'.format(number10, number11, number10 + number11))
print('{}'.format('안녕'))


#f_string
hour = 7
print(hour, type(hour))
# print('나는' + hour + '시에 밥을 먹을거야.') #TypeError 
print(f'나는{hour}시에 밥을 먹을거야')
```
