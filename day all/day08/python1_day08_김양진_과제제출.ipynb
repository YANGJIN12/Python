2. 딕셔너리형 성적 데이터로 평균 구하는 프로그램
records = { '짱구' : 72, '철수' : 100, '맹구' : 82, '훈이' : 63, '유리' : 91}
이렇게 학생 이름과 점수가 들어가 있을때 평균점을 계산해서 결과를 표시하고 성적 데이터의 목록과 평균과의 차이를 구하기
* 아래와 같은 결과가 출력되도록 할 것(왼쪽정렬 7문자, 오른쪽정렬 4문자, 오른쪽정렬 6문자)

[출력결과]
총점 : 408점
평균점수 : 81.6점 
유리       91   9.4
짱구       72  -9.6
철수      100  18.4
훈이       63 -18.6

```python
records = { '짱구' : 72, '철수' : 100, '맹구' : 82, '훈이' : 63, '유리' : 91}
# sum()메소드를 이용하여 딕셔너리 밸류값만 출력하는.values()메소드의 밸류값의 합을 구함
sum1 = sum(records.values())
print("총점 : {}점 ".format(sum1))
# 밸류값의 합에 / 5를 하여 평균값을 구함
mean = sum(records.values())/5
print("평균점수 : {}점".format(mean))
# .items()메소드로 키와 값을 분해해서 출력하고 for반복문을 사용해 각 키와 값을 한줄에 하나씩 출력하고
# .format() 메소드를 사용해서 Key, Value 값을 출력할때 각각 넣어줌
for k,v in records.items():
    print("{} {:>7d} {}".format(k, v, ))
```

4. 우리나라 전화번호는 '지역번호-국번-가입자개별번호' 형식으로 되어있다.
02-534-2425
02-4069-7534
031-3646-0796
032-691-0123

어떤 형식의 번호를 입력하더라도 '국번'을 추출혀여 출력하는 프로그램 구현하기

[출력결과]
전화번호를 입력하세요 : 02-1234-5678
1234

```python
 

def getMobile(s):
    header = ['010','011','016','017','018','019']
    ans = ''
    for i in range(len(s)):
        if s[i].isnumeric():
            ans = ans + s[i]
    if len(ans)==10 or len(ans)==11:
        if ans[:3] not in header:
            return False
    else:
        return False
    return ans

```

5. 커피자판기 프로그램 만들기. 커피자판기는 다음과 같이 동작한다.
(1) 커피자판기에 돈과 주문할 커피를 전달한다.
(2) 주문할 수 있는 커피의 종류와 가격은 아래와 같다.
		- 아메리카노 : 4100원
		- 카페라떼 : 4600원
		- 카푸치노 : 5000원
(3) 없는 커피를 주문할 경우 입력한 돈을 그대로 반환한다.
(4) 구매 금액이 부족하면 입력한 돈을 그대로 반환한다.
(5) 정상주문이면 주문한 커피와 잔돈을 반환한다.

[출력결과]
1. 없는 커피 주문할경우
커피를 선택하세요(아메리카노, 카페라떼, 카푸치노) : 콜드브루

2. 주문 금액 부족한 경우

3. 정상 주문인 경우
커피를 선택하세요(아메리카노, 카페라떼, 카푸치노) : 카페라떼
얼마를 내시나요 : 5000원
4000원에 카페라떼를 선택하셨습니다.
잔돈 400원. 커피 카페라떼

```python
menu = {'-아메리카노':4100,'-카페라떼':4600,'-카푸치노':5000}  #-없앰
def receipt():
    for k,v in menu.items():
        print("{} {:>7d}원 ".format(k, v))
#    print("{}원에 {}를 선택하셨습니다.".format(money,choice))
def  condition(choice,money): #품목과 돈을 인수로 받고
    if choice not in menu: 
        print("잔돈 {}원. 커피 없는메뉴".format(money))
    elif money < menu[choice]:
        print("잔돈 {}원. 커피 금액부족".format(money))
    else:
        print("잔돈 {}원. 커피 {}".format(money-menu[choice],choice))
receipt()
choice = input("커피를 선택하세요(아메리카노, 카페라떼, 카푸치노): ") #주문은 메뉴를 보여준 후에
money = int(input("얼마를 내시나요 : "))
print("{}원에 {}를 선택하셨습니다.".format(money,choice))
condition(choice,money) #품목과 돈을 인수로 넘긴다.
```
