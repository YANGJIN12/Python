1. 파이썬, 자바, C점수의 합계와 평균을 구한 뒤 결과 출력하기

```python
#1) 사용자로부터 파이썬 점수, 자바 점수, C 점수를 입력받는다. 예) 50.5
python = float(input("파이썬 점수: "))
java = float(input("자바 점수: "))
c = float(input("C 점수: "))

#2) 입력한 값들의 합계와 평균을 구한다.
sum = python + java + c
average = sum / 3
#3) 결과 출력한다. 예) 총 점수 합계는 **점, 총 평균점수는 **.**점
print('총 점수 합계는 {:.0f}점, 총 평균점수는 {:.2f}점'.format(sum, average))
print(f"총 점수 합계: {int(sum)}, 총 평균점수: {average:.2f}")
```

2. 복합대입연산자, 관계연산자 출력결과 확인하기

```python
#1) 정수형 변수 2개를 생성하고 각 변수에 값을 40과 2를 대입시킨다.
a = 40
b = 2
print(a,b)
#2) 복합 대입연산자를 모두 사용해 두개의 변수 출력결과를 확인한다.(출력결과 6개)
a += b
print(a)
a -= b
print(b)
a *= b
print(a)
a **= b 
print(b)
a /= b
print(a)
a %= b 
print(b)
#변수를 사용하는 이유 > 재할당
# 3)관계연산자를 사용해 두개의 변수 출력결과를 확인한다.(출력결과 6개)
print(a < b)
print(a > b)
print(a <= b)
print(a >= b)
print(a == b)
print(a != b)
```

4. 노래 한 곡당 300원인 코인 노래방에서 금액을 입력받고 부를 수 있는 노래와 잔돈 출력하기

```python
# [결과] 
#  금액 : 2000
#  부를 수 있는 노래는 6곡이고, 잔돈 200원이 반환되었습니다.

money = int(input("금액: "))
song = money // 300
change = money % 300
print('부를 수 있는 노래는 {}곡이고, 잔돈은 {}원이 반환되었습니다.'.format(song, change))
print(f"부를 수 있는 노래는 {song}곡이고, 잔돈은 {change}원이 반환되었습니다.")
```
