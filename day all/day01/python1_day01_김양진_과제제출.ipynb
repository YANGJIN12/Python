4. 출력결과가 아래와 같도록 만들기
(변수명은 height, weight로 작성 / 변수명 옆에 주석으로 키와 몸무게 표시해줄 것 / 기본자료형 int, float 사용)

```python
height = 180 # 키는 180cm
weight = 83.5 # 몸무게는 83.5kg
print("내 키는",int(height),"cm이고", "내 몸무게는", float(weight),"kg입니다.") 

print("내 키는 ",  end='')
print( int(height), end='')
print("cm이고", end='')
print(" 내 몸무게는 ", end='')
print(float(weight), end='')
print("kg입니다")

print("내키는 %d" %(height))
print("내키는 %f" %(weight)) 

```
