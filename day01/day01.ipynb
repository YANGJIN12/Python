```python
#반지름을 전달하면 원의 넓이를 반환하는 함수 구현
import math

#get_area() 함수 정의
def get_area(radius):
  '''반지름을 입력받아 원의 넓이를 반환하는 get_area()함수'''
  area = math.pi * math.pow(radius, 2)
  return area

radius = 1.5
area = get_area(radius)
print(area)
print(get_area.__doc__)
```

#3. 기본자료형(Data Type)

```python
#정수형 자료형(int)
num1 = 3
print(num1, type(num1))
num2 = 10
print(num2, type(num2))

#실수형 자료형(float)
num3 = 3.14
print(num3, type(num3))

num4 = 5.0
print(num4, type(num4))

#type() 함수 : ()안에 있는 변수명 또는 값들의 자료형을 알려주는 함수

#정수->실수형 형변환
print(float(num1), type(float(num1)))
print(float(num2), type(float(num2)))
print(float(num3), type(float(num3)))

#실수->정수형 형변환
print(int(num3), type(int(num3)))
print(int(num4), type(int(num4)))
```

```python
#변수명 int1, 값은 100
#변수명 float1, 값은 20.414
#각각 값과, 타입을 출력
#형변환 int->float, float->int 출력

int1 = 100
float1 = 20.414
print(int1, type(int1))
print(float1, type(float1))

print(float(int1), type(float(int1))) 
print(int(float1), type(int(float1)))
```
