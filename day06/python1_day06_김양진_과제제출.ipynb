1. fruits 객체의 리스트 안에 '사과', '바나나', '망고', '배', '복숭아', '감'이 들어있다.
'배'가 fruits 안에 들어있다면 지우고, 없다면 '배는 fruits 안에 없습니다.'라는 문장 출력하기(조건문, in연산자, 메소드 사용)

```python
fruits = ['사과','바나나','망고','배','복숭아','감']
if '배' in fruits:
    fruits.remove('배')
    print(fruits)
else:
    print("배는 {} 안에 없습니다.".format(fruits))
```

11. 다섯개의 숫자를 입력받아 리스트에 저장한 후 입력된 반대 순서대로 출력하는 프로그램

```python
my_list = []
for i in range(5):
    num = int(input("숫자 입력:"))
    my_list.append(num)
    my_list.reverse()
    print(my_list)
```
