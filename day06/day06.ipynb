# 문제2
사용자에게 정수 5개를 입력받아서 총합과 평균 구하기(메소드,for,len()이용)


```python
#1. 빈리스트 만들기
#2. 변수설정
#3. for문 안에서 값을 입력받도록
#4.빈리스트에 값을 추가
#5.변수에 +
#6. 총합, 평균 출력
bin_list = list()
sum = 0

for i in range(5):
    num = int(input("정수 입력:"))
    bin_list.append(num)
    sum += bin_list[i]
print(f'총합: {sum}, 평균: {int(sum / len(bin_list))}')


```

# 2. 튜플(tuple)

```python
tuple1 = (1,1,2,3,3,4,4,5,5,6,7,7,7,7,9)
print(tuple1, type(tuple1))

# 인덱싱 / 슬라이싱
print(tuple1[1])
print(tuple1[3:6])

#  원소 조회 가능 in, not in
print(10 in tuple1)
print(7 not in tuple1)

# .count() 개수를 세는 메소드
print(tuple1.count(7))
print(tuple1.count(4))

tuple2 = ('a','b','c','c','a','b','b','b','b','b')
print(tuple2.count('c'))
# .index(값) 몇번째 인덱스에 속하는지 알려주는 메소드
print(tuple2.index('c'))
print(tuple2.index('b')) # 맨 앞에 나온 요소의 인덱스를 알려줌

# tuple3 = ((100,30,45,77,99))
# print(max(tuple3)) # max() 최대값 함수
# print(min(tuple3)) # min() 최소값 함수
# print(sum(tuple3)) # sum() 요소값들의 합
# print(sum(tuple3)/len(tuple3)) # 평균값 함수

# 읽기전용 리스트
# 추가, 수정, 삭제 > 리스트에서 가능한 메소드 존재
#tuple3.append(20) # AttributeError: 'tuple' object has no attribute 'insert'
print(tuple3)
#tuple3.insert(0,55) # AttributeError: 'tuple' object has no attribute 'insert'
print(tuple3)
#tuple3.pop()   # AttributeError: 'tuple' object has no attribute 'insert'
print(tuple3)

tuple4 = list(tuple3)
print(tuple4)
tuple4.append(60)
print(tuple4)
tuple3 = tuple(tuple4)
print(tuple3)
```

# 3. 세트

```python
set1 = {'사과','참외','멜론','복숭아'}
print(set1, type(set1))

# .add() 추가
set1.add('포도')
print(set1)
set1.add('파인애플')
print(set1)
set1.add('사과') # 중복된 값은 추가를 해도 추가되지 않음
print(set1)

# .remove() 삭제
set1.remove('멜론')
print(set1)
#set1.remove('멜론')    # KeyError: '멜론' > 없는 요소를 삭제하려고 하면 오류 발생
#print(set1)

# .discard(값)
set1.discard('복숭아')
print(set1)
set1.discard('복숭아')
print(set1)

# .pop()
#set1.pop(0) # TypeError: pop() takes no arguments (1 given) > 인덱스를 넣게되면 오류 발생
#print(set1) # 요소 한개를 삭제

set1 = {'사과','참외','멜론','복숭아'}
set2 = {'참외', '복숭아','포도'}
# 교집합 &
print(set1 & set2)
# 합집합 | > 중복된 내용은 한번만 출력
print(set1 | set2)
# 차집합 -
print(set1 - set2)
print(set2 - set1)
```
