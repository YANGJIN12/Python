```python
# 과제2. 멜론 TOP 100가수명, 노래제목 추출 후 파일명:melon.csv으로 저장하기
!pip install pandas
!pip install Selenium
import csv
from bs4 import BeautifulSoup as bs
import pandas as pd
from google.colab import data_table

song_list = [] # 노래제목(텍스트) 리스트
singer_list = [] # 가수명(텍스트) 리스트
rank_list = [] # 순위 리스트

header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
res= requests.get("https://www.melon.com/chart/index.htm",headers = header)
soup = bs(response.text,'lxml')
print(response.status_code)
singer = soup.select("span.checkEllipsis")
for i in singer:
    print(i.text)
song = soup.select("div.rank01>span")
for i in song:
    print(i.text.strip())


for i in range(len(song)):
    song_list.append(song[i].text.strip())
    singer_list.append(singer[i].text.strip())
    rank_list.append(i+1)
    print(i.text)

info = {"rank" : rank_list,"singer":singer_list, "title":song_list}

music = pd.DataFrame(info)
music.set_index('rank', inplace = True )
music.to_csv("melon.csv", 'w', encoding = "utf-8")


```
