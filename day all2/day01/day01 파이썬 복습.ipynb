#4) 조건문

```python
# if
# if~elif~else
# if~else

# 나이를 입력받아서 100살이면 초고령, 65살 고령, 40살 이상 중년, 20살이상 청년, 그외에 미성년
age = int(input("나이 입력: "))
if age >= 100:
    print("초고령입니다.")
elif age >= 65:
    print("고령입니다.")
elif age >= 40:
    print("중년입니다.")
elif age >= 20:
    print("청년입니다.")
else:
    print("미성년입니다.")

# 80점 이상일 때 합격,  100점일 경우에는 만점으로 합격
score = int(input("점수 입력: "))
if score >= 80:
    if score == 100:
        print("만점으로 합격입니다.")
    else:
        print("합격입니다.")
else:
    print("불합격입니다.")
```

```python
# while문으로 0부터 10까지 출력하는 프로그램
num = 0
while num <= 10:
    print(num)
    num += 1


# 빈리스트에 반복문을 통해 값을 추가(range함수사용, append메소드이용,1, 11)
list1 = []
for i in range(1,12):
    list1.append(i)
print(list1)

# 리스트 컴프리헨션 1부터 30까지 홀수만 출력(range함수이용, 시작값과 종료값 설정, if문 사용)
num2 = [i for i in range(1,31) if i % 2 == 1]
print(num2)

num3 = [i for i in range(100) if i % 2 == 0]
print(num3)
```

# 5) 사용자 정의 함수

```python
# 함수명 getsum, 매개변수 1개
# 1부터 호출받은 숫자까지 총합을 구해서 return 해주는 함수
# 호출받을 때 인수에 3을 넣었을 때 결과 출력
def getsum(num):
    total = 0
    for i in range(1, num+1):
        total += i
    return total

sum_t = getsum(3)
print(sum_t)


```
