```python
!pip install Selenium
!apt-get update
!apt install chromium-chromedriver
import requests
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options) # wd라는 객체에 크롬 드라이버 넣어줌
# 드라이버가 찾아서 들어온것으로 생각 설치해줘야하는 경우가 있으면 드라이버 설치해주면됨
wd.get("https://vibe.naver.com/chart/total")
#.get url에 접근하는 메소드

# 네이버 뮤직에서 크롤링
html = wd.page_source
soup = bs(html, 'html.parser')

# 순위, 가수명, 노래제목 리스트형태로 출력
song_box = soup.find('tbody')
songs = song_box.select('td.song')

song_rank = []
rank = 1
for i in songs:
    ranks = str(rank) + "위"
    title = i.select_one('a.link_text').get('title')
    # select_one은 가장 처음의 <a>태그를 찾고 바로 속성값을 얻을 수 있음
    # a태그 내 class link_text를 찾고 .get('title')로 title태그의 내용을 불러옴
    singers = i.select_one('a.link_artist').text
    # a태그 내 link_artist 클래스를 찾고 텍스트만 불러온다
    song_rank.append([ranks, title, singers])
    # song_rank라는 빈리스트에 append 메소드를 이용해 ranks, title, singers를 추가함
    rank += 1
print(song_rank)
```

```python
import numpy as np

# 넘파이 배열을 만들 때 np.array([])
ar1 = np.array([1,2,3,4,'a','b'])
ar2 = np.array([[1, 2], [3, 4]])
ar3 = np.array([[5, 6], [7, 8]])
print(ar1, type(ar1))
print(type(ar1.dtype))

# 정수와 실수 > 실수
# 정수와 문자열 > 문자열
# 실수와 문자열 > 문자열

print(ar2 + ar3)
```

# 4. 넘파이

## 4-3) 넘파이 속성

```python
import numpy as np

# 1차원 배열
arr1 = np.array([1.0, 20])
print(arr1, type(arr1))
print(arr1[1])

# 2차원 배열(행과 열이 생김)
arr2 = np.array([[1,2,3], [4,5,6]])
print(arr2)
print(arr2[0][1])

# 크기 .shape
print(arr1.shape)
print(arr2.shape)

# 차원 .ndim
print(arr1.ndim)
print(arr2.ndim)

# 총 원소 개수 size
print(arr1.size)
print(arr2.size)

# 원소들의 데이터 타입
print(arr1.dtype)
print(arr2.dtype)
```

## 4-4) 넘파이 함수

```python
# 넘파이 검색, 수정, 추가, 삭제
arr1 = np.array([1.0, 20])
arr2 = np.array([[1,2,3], [4,5,6]])

# 검색(추출, 색인)
print(arr1[1])
print(arr2[1][1])
print(arr2[0][2])
print(arr2[0])

# 수정(기존에 있는 값이 변경된 값으로 변수에 재할당)
print(arr1[0])
# arr1[0] > 10으로 값 변경
arr1[0] = 10
print(arr1[0])

# arr2에서 0행 1열을 7로 변경
arr2[0][1] = 7
print(arr2)

# 추가(변수에 재할당)
arr3 = np.array([[0, 0], [0, 0]])
print(arr3)
print(arr3.ndim)

# 행 추가
arr3 = np.append(arr3, [[1, 1]], axis = 0) # 배열 그대로 행 추가
print(arr3)

# 열 추가
arr4 = np.append(arr3, [[1,1], [2,2], [3,3]], axis = 1) # axis = 1 뒷쪽에 추가됨을 확인
print(arr4)

# 삭제
# np.delete(배열, 인덱스, 축)
# axis = 0 > 행,  axis 1 > 열(2차원)
# axis = 0 > 높이, axis 1 > 행, axis = 2 > 열(3차원)

print(np.delete(arr4, 1, axis = 0)) # 2행 삭제
print(np.delete(arr4, 1, axis = 1)) # 2열 삭제

```
