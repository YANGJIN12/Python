# 1. 자료구조

## 1-2) 추상 데이터 자료형

```python
# 가방(Bag())의 추상 자료형 정의 및 구현
# data > 지갑, 핸드폰, 거울, 빗, 동전 등
# 클래스 형태
class Bag:
    def __init__(self):
        self.mybag = []

    def contains(self, bag, e):
        return e in bag

    def insert(self, e):
        self.mybag.append(e)

    def remove(self, bag, e):
        self.mybag.remove(e)

    def count(self, bag):
        return len(bag)

# 객체화
b1 = Bag()
# 핸드폰, 지갑, 빗, 손수건
b1.insert("핸드폰")
b1.insert("지갑")
b1.insert("빗")
b1.insert("손수건")
print(f"가방속 물건: {b1.mybag}")
# 가방속의 물건 개수 확인
num = b1.count(b1.mybag)
print(f"가방속 물건 개수: {num}")
# 중복삽입 > 빗
b1.insert("빗")
print(f"가방속 물건: {b1.mybag}")
# 가방속에 지갑이 있는지 여부 contains
result = b1.contains(b1.mybag, "동전")
print(f"가방속 지갑이 있는지 여부: {result}")
#핸드폰을 꺼냄
b1.remove(b1.mybag, "핸드폰") # 객체명.메소드명(메게변수2개(위치, 빼낼 것))
print(f"가방속 물건: {b1.mybag}")
```

# 2. 알고리즘 성능 분석

```python
# append, insert 시간차이
import time
start = time.time()
class Bag1():
    def __init__(self):
        self.new_bag = []
    def append(self, e):
        self.new_bag.append(e)


b2 = Bag1()
b2.append("축구공") 
end = time.time()
print(f"실행시간: {end-start}")
# 실행시간 확인 시 초단위 출력 append가 insert보다 빠른 실행시간을 가짐
```

```python
import time
start = time.time()
class Bag1():
    def __init__(self):
        self.new_bag = []    
    def insert(self, e):
        self.new_bag.insert(0, e)

b2 = Bag1()
b2.insert("축구공") 
end = time.time()
print(f"실행시간: {end-start}")
# 실행시간 초단위
```

## 2-1) 시간복잡도

```python
# O(1)
def hello_world():
    print("hello world")
```

```python
# O(n)
def print_each(li):
    for i in li:
        print(i)
```

```python
# O(n^2)
def print_each2(li):
    for i in li:
        for j in li:
            print(i, j)
```

> O(log n), O(n log n) : 정렬 알고리즘에서 많이 사용됨

# 4. 연결리스트

## 1) 단일연결리스트

```python
# 노드 생성
class Node:
    def __init__(self):
        self.data = None
        self.link = None

node1 = Node()
node1.data = "짱구"
# print(node1.data)

# 노드 연결
node2 = Node()
node2.data = "철수"
# 링크 연결 node1 > node2 (짱구 > 철수)
node1.link = node2
# print(node1.data)
# print(node1.link)
# print(node2.data)

# node3 훈이
node3 = Node()
node3.data = "훈이"
node2.link = node3

# node4 맹구
node4 = Node()
node4.data = "맹구"
node3.link = node4 

# # 노드 삭제 node3을 지울때
# print(id(node3.link))
# node2.link = node3.link # node2.link 기존에 훈이를 가르킴, node3.link 기존에 맹구로 변경하면됨
# print(id(node2.link))
# # del(node3)
# # print(node3)
```

```python
# 1. 첫번째 노드를 현재(current) 노드로 지정하고 현재 노드의 데이터 짱구를 출력함
current = node1 
print(current.data, end = '->')
# 2. 현재 노드의 링크가 비어있지 않다면 현재 노드를 현재 노드의 링크가 가리키는 노드로 변경한 후에 현재 노드의 데이터인 철수를 출력함
while current.link != None:
    current = current.link
    print(current.data, end = '->')
# 3. 현재 노드의 링크가 비어있으면 종료함
while current.link == None:
    break
```
