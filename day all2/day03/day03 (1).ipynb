## 2) 예외처리 구문 활용

```python
# 제수: 나눗셈에서 어떤 수를 나누는 수 
# 피제수: 어떤수를 다른 수로 나눌때 처음의 수 
# 제수와 피제수를 입력받아 0으로 나눌 수 없읍니다. 정수만 입력할 수 있습니다. 알수없는 예외가 발생했습니다.
try:
    num1 = int(input("제수 입력: "))
    num2 = int(input("피제수 입력: "))
    print(f"{num1} / {num2} = {num1 / num2}")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except ValueError:
    print("정수만 입력할 수 있습니다.")
except Exception:
    print("알 수 없는 예외가 발생했습니다.")
```

```python
print("프로그램 시작")
# arr = ['h','e','l','l','o']
# # print(arr[0])
# print(arr[8])
try:
    arr = ['h','e','l','l','o']
    # print(arr[0]) # 정상실행
    print(arr[8]) # 오류발생 > indexerror
    print("try문 실행") 
except:
    print("except문 실행")
else:
    print("else문 실행")
finally:
    print("finally문 실행")

print("프로그램 종료")
```

```python
# raise
num = int(input("1~5까지 숫자입력: "))

# 1보다 작거나 5보다 크면 에러발생
if num < 1 or num > 5:
    raise # RuntimeError: No active exception to reraise > 강제로 에러를 발생해 프로그램이 죽는 것을 의미
# 범위 안에 있으면 정상 출력
print(f"입력한 값: {num}입니다.")
```

```python
# raise + 예외처리 이름
num = int(input("1~5까지 숫자입력: "))

# 1보다 작거나 5보다 크면 에러발생
if num < 1 or num > 5:
    raise ValueError
# 범위 안에 있으면 정상 출력
print(f"입력한 값: {num}입니다.")
```

```python
# raise + 예외처리 이름 + 문장
num = int(input("1~5까지 숫자입력: "))

# 1보다 작거나 5보다 크면 에러발생
if num < 1 or num > 5:
    raise Exception("에러에러에러!")
# 범위 안에 있으면 정상 출력
print(f"입력한 값: {num}입니다.")
```

```python
# try-except
try:
    num = int(input("1~5까지 숫자입력: "))

# 1보다 작거나 5보다 크면 에러발생
    if num < 1 or num > 5:
        raise
# 범위 안에 있으면 정상 출력
    print(f"입력한 값: {num}입니다.")
except:
    print("1~5 사이의 값만 입력하세요")
```
