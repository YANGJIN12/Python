# 1. find() 메소드 사용

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.naver.com'

response = requests.get(url)
soup = bs(response.text, 'html.parser')
# print(soup)

# 1. 태그 이름 지정
div = soup.find('div') # find('div') div 태그 중에서 태그명이 div 첫번째 것만 추출
print(div, type(div))

# div1 = soup.find_all('div') # find_all 메소드를 사용하면 div라는 태그명을 가진 모든 내용 추출
# print(div1, type(div1))

# 2. 태그 이름을 지정해서 내용 출력
div = soup.find('div').text
print(div)

# 3. 태그 속성만 지정
div = soup.find('div', class_ = 'title_area')
print(div)
print(div.text)

# 예제 오늘 읽을 만한 글만 추출
# 태그 속성을 지정
div = soup.find('strong', class_ = 'title')
print(div.text)

# 4. 속성값 이용
div = soup.find('ul').get('class')
# ul태그가 있는 첫번째를 찾아서 class명을 찾는 것
# 클래스명이 리스트 형태로 나옴 
print(div)
 

# div = soup.find_all('div', class_ = 'group_nav').text
div = soup.find_all('div', class_ = 'group_nav')
for i in div:
    print(i.text)

div = soup.find('strong', class_ = 'title')
print(div.text)
```

> find_all을 이용하게되면 바로 text형태로 추출 불가

> <에러내용> AttributeError: ResultSet object has no attribute 'text'. You're probably treating a list of items like a single item. Did you call find_all() when you meant to call find()?

> <발생이유> 리스트 형태로 출력되기 때문에 text를 사용하지 못함

> 1) find_all 사용 시 인덱스 번호를 지정
> 2) find로 사용

# 2.멜론사이트 html코드 확인
