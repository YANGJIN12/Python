```python
import random

print(random.randint(1,5)) # randint(a, b) > a이상 b이하의 숫자중에서 랜덤한 숫자를 return

random.randrange(10) # randrange(a) > 0~a-1 사이의 숫자중 랜덤한 숫자를 return
random.randrange(2, 10, 3) # randrange(a, b, c) > a, a+c, a+c+c, ... 숫자 중 랜덤한 숫자를 return

print(random.choice("안녕하세요")) #choice(시퀀스 자료형) > 순서가 있는 자료형(리스트, 튜플, 문자열, range())

print(random.sample(['봄','여름','가을','겨울'], 2)) # sample(시퀀스 자료형, 개수) > 해당 시퀀스 자료형에서 개수만큼 요소를 뽑아줌
```

# 2. 사용자 정의 함수

```python
# 1부터 매개변수로 받은 숫자까지 총합을 구해서 return 해주는 함수
# 함수명 getSum(num)

def getSum(num):
    sum = 0
    for i in range(1, num + 1):
        sum += i
    
    return sum 

number = getSum(3)
print(number)

```

# 3. 클래스

```python
num1 = 3 # num1 변수, int형 객체
num2 = 5.5 # num2 변수, float형 객체
print(num1, type(num1), num2, type(num2))
print(float(num1), type(float(num1))) # int 타입으로 생성된 객체 num1을, float 클래스 생성자를 통해서 float 타입의 객체로 생성됨

list1 = list(range(3)) # list1 이라는 변수는 list타입으로 생성된 객체 list1
print(list1, type(list1))
# 튜플로 형변환
print(tuple(list1), type(tuple(list1))) # list 타입으로 생성된 객체 list1을 tuple 클래스 생성자를 통해 tuple 타입의 객체로 생성됨

list1.append(3) # 객체명.메서드명
print(list1)
```

> 클래스를 사용하는 이유

변수 = 값
여러값을 저장하고 싶을 때 > 새로운 저장공간이 필요 > 컬렉션(list, tuple, set, dict)

여러값들과 기능을 저장하고 싶을 때 > 새로운 저장공간이 필요 > 사용자 정의 함수

사용자 정의 함수를 여러개 저장하고 싶을때 > 새로운 저장공간이 필요 > 클래스

```python
# 클래스 선언
# class 클래스명:
#     def 메서드명(self, 매개변수, ...):
#         self.변수 = ...
#         return 반환식

class Person: # 클래스 정의
    def getName(self): # 메서드 정의(사람의 이름을 정의)
    # self는 인스턴스 자신을 의미함, 인스턴스 내부의 속성을 다루기 위해 "자신의 속성을 다룸"을 알려주기 위해서 self를 사용함
        return self.name  # name 데이터 속성 정의(사람의 이름을 나타냄)

    def getAge(self): # 메서드 정의(사람의 나이를 정의)
        return self.age # age 데이터 속성 정의(사람의 나이를 나타냄)

# 메서드 내부에서 self.필드명으로 접근하면 여러 객체도 한번의 선언으로 구별해서 다룰 수 있게됨

# 클래스 사용
# [객체화]
# 객체명 = 클래스명()
# [각 필드에 접근]
# 객체명.메서드명()

pr = Person() # 객체화 / 인스턴스 작성
pr.name = "김양진" # 필드에 접근, 데이터 속성에 값을 대입, 현재 Person에는 아무런 속성값이 없으므로 [객체명.속성]으로 추가가능
pr.age = 22 # 데이터 속성에 값을 대입
name = pr.getName()
age = pr.getAge()
print(f"{name}의 나이는 {age}입니다.")
```

```python
class Cal:
    def setdata(self, first, second):
        self.first = first
        self.second = second


    def add(self):
       return self.first + self.second

    def mul(self):
        return self.first * self.second

num1 = Cal()
num1.setdata(5, 5)
print(num1.add(), num1.mul())


```

# 4. 생성자

```python
# 축구선수 1명과 야구선수 1명의 이름과 나이 직업을 넣는 클래스
class SomeClass:
    def __init__(self, name, age, job):
        self.name = name
        self.age = age
        self.job = job

    def introduce(self):
        print(self.name)
        print(self.age)
        print(self.job)

# 생성자를 사용해서 인스턴스의 생성과 초기값 저장을 한번에 가능하다.
user1 = SomeClass('류현진', 36, '야구선수')
user2 = SomeClass('손흥민', 31, '축구선수')

user1.introduce()
user2.introduce()

# 객체별 속성을 추가할 수 있음
user2.country = '영국'
print(user2.country)

print(user1.country)
```

> 클래스의 장정 : 1명이 아니라 많은 사람들의 정보가 필요하다면 굳이 한명씩
이름, 성별, 나이에 해당하는 변수를 생성할 필요없이 클래스를 이용해서 간편하게
만들 수 있음

#5. 상속

```python
class Person:
    def __init__(self, name):
        self.name = name

    def eat(self, food):
        print(f"{self.name}가 {food}를 먹는다.")

class Student(Person):
    def __init__(self, name, school):
        super().__init__(name)
        self.school = school

    def study(self):
        print(f"{self.name}는 {self.school}에서 공부함")

# 객체화는 서브클래스를 바탕으로 객체화함
potter = Student('김양진', '경동고') # 객체화
# 객체.서브클래스 메서드() > 호출 잘됨
potter.study()
# 객체.슈퍼클래스 메서드() > 호출 잘됨
potter.eat('밥')

use = Person("짱구")
use.eat('초코과자')

# use.study() > 슈퍼클래스의 객체로 서브클래스의 메서드를 호출하는 것은 불가능함
```

```python
# 서브클래스의 인스턴스 자료형
# 슈퍼클래스의 객체는 슈퍼클래스의 인스턴스
# 서브클래스의 객체는 서브클래스의 인스턴스이면서 슈퍼클래스이 인스턴스

print(isinstance(potter, Person))
print(isinstance(potter, Student))
```
