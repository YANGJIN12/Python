1. 로또 추첨기 만들기
1~45까지 수 중에 랜덤한 6개의 숫자를 뽑아(중복되지 않게) 오름차순과 내림차순으로 정렬하기(리스트형태로 출력)

```python
import random as rd

lotto = []
rnd_num = rd.randint(1, 45)

for i in range(6):
    while rnd_num in lotto:
        rnd_num = rd.randint(1, 45)
    lotto.append(rnd_num)

print("로또번호: {}".format(sorted(lotto)))
print("로또번호: {}".format(sorted(lotto, reverse=True)))

```

2. 시험 점수를 관리하는 프로그램(학생번호, 이름, 점수를 관리) 
각각의 학생을 나타내는 클래스, 인스턴스 생성은 4명까지 
평균 점수 출력하기
- 클래스명 : Student
- 초기화 메서드 생성(매개변수는 self, id, name, score=0)
- id를 얻어낼 메서드
- 학생이름을 얻어낼 메서드
- 점수를 얻어낼 메서드
- 점수 저장하는 메서드

- 클래스명 :CalcScore
- 초기화 메서드 생성(매개변수는 self, student)
- 학생 추가하는 메서드
- 평균 구하는 메서드

- 객체 생성


[결과화면]
평균 점수 = 생성한 4개 객체의 평균 점수

```python
class Student:
    def __init__(self, id, name, score=0):
        self.id = id 
        self.name = name
        self.score = score
    
    def ID(self):
        return self.id

    def Name(self):
        return self.name

    def Score(self, score):
        self.score = score
     
    def Score(self):
        return self.score

class CalcScore:
    def __init__(self):
        self.students = []

    def Plus_Student(self, student):
        self.students.append(student)
    
    def mean_score(self):
        total = 0
        for i in self.students:
            total += i.Score()
        ave_v = total / len(self.students)
        return ave_v

p1 = Student(1, "짱구", score = 51)
p2 = Student(2, "철수", score = 100)
p3 = Student(3, "훈이", score = 78)
p4 = Student(4, "맹구", score = 82)

calc = CalcScore()
calc.Plus_Student(p1)
calc.Plus_Student(p2)
calc.Plus_Student(p3)
calc.Plus_Student(p4)
print(f"평균점수  = {calc.mean_score()}")

    
```
