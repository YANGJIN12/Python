## 2) 클래스로 구현

> create(size) : 큐를 생성
> is_empty() : 비어있다면 True, 아니면 False
> if_full() : 가득 찼다면 True, 아니면 False
> enqueue() : 큐에 데이터 추가(rear)
> dequeue() : 큐에 데이터 삭제(front)
> 

```python
class Queue:
    def __init__(self, max = 5):
        self.que = [None for i in range(max)]
        self.max = max
        self.rear = 0

    def isEmpty(self):
        return self.rear == 0

    def isFull(self):
        return self.max == self.rear

    # 데이터 삽입
    def enqueue(self, element):
        if not self.isFull():
            self.que[self.rear] = element
            self.rear += 1
    
    def dequeue(self):
        if not self.isEmpty():
             temp = self.que[0]
             for i in range(self.rear - 1):
                 self.que[i] = self.que[i + 1]
             self.rear -= 1
    # 디큐를 진행하면 한칸씩 앞으로 당겨지고 값에 None으로 들어가야하기 때문에 rear에 -1을 진행함
             for j in range(self.rear, self.max):
                 self.que[j] = None
             return temp

    def peek(self):
        if not self.inEmpty():
            return self.que[0]
    
    def clear(self):
        self.rear = 0

    def show(self):
        for i in range(self.rear):
            print(self.que[i], end = ' | ')
        print()
```

```python
# 객체화
que = Queue()
que.enqueue(10)
que.enqueue(20)
que.enqueue(30)
que.show()

print("삭제: ", que.dequeue())
que.show()

print(que.que)

que.enqueue(100)
que.enqueue(200)
que.enqueue(300)
que.show()
print(que.que)

# 큐의 길이를 지정했기 때문에 큐의 길이를 벗어난 값을 넣을 수는 없다
que.enqueue(400)
que.show()
print(que.que)

que.clear()
que.show()
print(que.que)
```

# 2. 원형큐(Lingbuffer)

```python
class Circleque:
    def __init__(self, max):
        self.que = [None] * max
        self.length = self.front = self.rear = 0
        # 요소 개수   # 삭제될 요소 idx # 추가될 idx
        self.max = max

    def isEmpty(self):
        return self.length == 0

    def isFull(self):
        return self.length == self.max

    def enqueue(self, element):
        if not self.isFull():
            self.que[self.rear] = element # rear 포인터에 입력받은 인수값을 넣는다
            self.rear = (self.rear + 1 ) % self.max # 전체길이로 나머지 연산을 하여 # 포인터가 전체길이를 넘지 못하게 이동시킴
            self.length += 1

    def dequeue(self):
        if not self.isEmpty():
            temp = self.que[self.front]
            self.front = (self.front + 1) % self.max
            self.length -= 1
            return temp
    
    def peek(self):
        if not self.isEmpty():
            return self.que[self.front]  # 배열의 요소 반환

    def clear(self):
        self.front = self.rear
        self.length = 0

    def show(self):
        print(f"front : {self.front} rear = {self.rear}, length : {self.length}")
        idx = self.front
        for i in range(self.length):
            print(self.que[idx], end = ' | ')
            idx = (idx + 1) % self.max
        print()

    print()
```

```python
c = Circleque(5)
c.show()

c.enqueue(10)
c.enqueue(20)
c.enqueue(30)
c.enqueue(40)
c.show()

c.dequeue()
c.show()

c.enqueue(100)
c.enqueue(200)
c.show()

c.dequeue()
c.show()

```

# 3. 정렬(select sort)

## 3-1) 선택정렬

```python
def find_select(arr):
    minnum = 0
    for i in range(1, len(arr)):
        if(arr[minnum] > arr[i]):
            minnum = i
    return minnum

before = [11,22,20,19,50,38]
after = []

print(f"정렬전 : {before}")

# 정렬 시키기 위한 반복문
for j in range(len(before)):
    minli = find_select(before)
    after.append(before[minli])
    del(before[minli])
print(f"정렬후 : {after}")
```

```python
def select_find(arr):
    n = len(arr)

    # 이중 for문
    for i in range(0, n - 1): # 바깥 for문으로 지정해서 배열크기 -1 만큼 반복
        minnum = i
        for j in range(i + 1, n): # 안쪽 for문으로 지정해서 배열크기만큼 반복
            if(arr[minnum] > arr[j]):
                minnum = j
        tmp = arr[i]
        arr[i] = arr[minnum]
        arr[minnum] = tmp
        print(arr[:i + 1])
    return arr

data = [11,365,5,40,104]
print(f"정렬전 : {data}")
data = select_find(data)
print(f"정렬후 : {data}")
```

## 3-2) 삽입정렬(Insert sort)

```python
def insert_find(arr):
    n = len(arr)
    for i in range(1, n): # i변수는 각 사이클의 끝 위치 지정
        for j in range(i, 0, -1): # 맨뒤부터 첫번째까지 반복
            if(arr[j-1] > arr[j]):
                arr[j-1], arr[j] = arr[j], arr[j-1]
        print(arr[:i + 1])
    return arr

data = [20, 1, 17, 5, 13, 9]
print(f"정렬전 : {data}")
data = insert_find(data)
print(f"정렬후 : {data}")
```

## 3-3) 버블정렬

```python
def bublle_find(arr):
    n = len(arr)
    # 패스과정(요소를 비교하고 교환하는 과정) n - 1번 실행
    for i in range(n-1, 0, -1):
        for j in range(0, i):
            if (arr[j] > arr[j+1]):
                arr[j], arr[j+1] = arr[j+1], arr[j]
        print(arr[:i+1])
    return arr

data = [20,1,17,5,13,9]

print(f"정렬전 : {data}")
data = bublle_find(data)
print(f"정렬후 : {data}")
```

# 4. 트리(Tree)

```python
# 트리 선언방법 1
class Tree:
    def __init__(self):
        self.val = None
        self.left = None
        self.right = None

tree = Tree()
print(tree)
tree.val = 30
print(tree.val)
tree.left = 15
tree.right = 25
print(tree.left, tree.right)
```

```python
# 트리 선언방법 2
class Tree:
    def __init__(self, val = None):
        if val != None:
            self.val = val
        else:
            self.val = None
        self.left = None
        self.right = None

tree = Tree('A')
print(tree.val)
tree.left = Tree('B')
tree.right = Tree('C')
print(tree.left.val, tree.right.val)
```
