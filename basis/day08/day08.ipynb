#1. 문자열 메소드

1) .format()

```python
print("10자리 폭 왼쪽정렬: '{:*<10d}'".format(123))
print("10자리 폭 오른쪽정렬: '{:*>10d}'".format(123))
print("10자리 폭 가운데정렬: '{:*^10d}'".format(123))
print("10자리 폭 왼쪽정렬: '{:<10d}'".format(123))
print("10자리 폭 오른쪽정렬: '{:>10d}'".format(123))
print("10자리 폭 가운데정렬: '{:^10d}'".format(123))
```

2) .count()

```python
string = '내가 그린 기린 그림은 목 긴 기린 그림이고, 네가 그린 기린 그림은 목 짧은 기린 그림이다.'
print(string.count('기린'))

string1 = 'best of best'
print(string.count('best'))
print(string1.count('best', 5))
print(string1.count('best', -7))
```

3) .find() : 앞에서 뒤로 찾고, 인덱스를 알려줌(첫번째)
   .rfind() : 뒤에서 앞으로 찾고, 인덱스를 알려줌(첫번째)

```python
a = 'apple'
print(a.find('p'))
print(a.find('z'))

string1 = 'best of best'
print(string1.find('best'))
print(string1.find('best', 5))
print(string1.rfind('best'))
```

6) .join()

```python
print('-'.join('python'))
# 반복가능객체 > 시퀀스 자료형,  문자열, 리스트, 튜플
# 딕셔너리 가능 > 키만 이용해서 가능함

print('+'.join(['p','y','t','h','o','n']))
print(''.join(['x','y','z']))
#print(''.join(['a':'apple', 'b':'banana'])) # SyntaxError: invalid syntax 키 값만 가능
```

7) .split()

```python
string = 'Life is too short'
print(string.split()) # 아무값이 없을 때는 공백을 기준으로 문자열을 나눠준다

number = '010-1234-5678'
print(number.split('.'))
number = '01.0-12.34-5.67.8'
print(number.split('.'))
```

8) .replace()

```python
print(string)
# short > long
print(string.replace('short','long')) # 글자 바꾸기 기능과 동일 콤마를 기준으로 앞에는 기존 문자열 뒤에는 변경할 문자열

print(number)
# - > 공백없이(붙여서)
print(number.replace('-',''))
print(number.replace('.', '가'))
```

9) 불필요한 문자열 제거 메소드

```python
a = '123*******'
b = '*******123'
c = '***123****'
print(len(a), len(a.rstrip('*'))) # 전체길이, 공백을 제거한 숫자
print(len(b), len(b.lstrip('*')))
print(len(c), len(c.strip('*')))
print(c.strip()) # 공백 or 특정문자열을 넣었을 때 제거하고 남은 값만 출력 # 중간에 공백 or 특정 문자열이 있을 때는 제거되지 않음
```

2) 디폴트 매개변수

```python
# 함수 정의
def greet(message = 'hello'): # 디폴트값 설정
    print(message)

# 함수 호출
# greet('안녕하세요') # TypeError: greet() missing 1 required positional argument: 'message' > 매개변수가 존재하는 인수가 들어가지 않아서 발생
greet()
greet("HELLO")
greet("반갑습니다.")
# greet('반갑습니다.','python', 'day', 'good', 123) # TypeError: greet() takes from 0 to 1 positional arguments but 5 were given > 인수가 여러개 들어감
```

3) 가변 매개변수

```python
# 함수정의 show()
def show(*args):
    print('arges : ', args)
    print('args의 타입 : ', type(args))
    for i in args:
        print(i)

# 함수 호출
show('python')
show('python', 'day', 'good', 123) # 인수가 여러개 들어가도 오류 미발생
```

4) 함수 호출(4가지 방법)

```python
# 함수 정의1 (매개변수 x, 리턴값 x)
def welcom():
    print('Hello World')

# 함수 호출1 > 함수 이름()
welcom()
print('호출 끝')

# 함수 정의2 (매개변수0, 리턴값x)
def print_name(num):
    print(num)
    for i in range(num):
        print("김양진")

# 함수 호출2 > 함수이름(인수)
print_name(3)

# 함수 정의3 (매개변수x, 리턴값0)
def get_my_name():
    name = '김양진'
    return name

#함수 호출 3 > 변수 = 함수이름()
my_name = get_my_name()
print(my_name)
print(get_my_name())

# 함수 정의4 (매개변수0, 리턴값0)
def sum_num(num1, num2):
    print(f"num1 = {num1}")
    print(f"num2 = {num2}")
    sum_result = num1 + num2
    return sum_result

# 함수 호출4 > 변수 = 함수이름(인수1, 인수2) #print(함수이름(인수1, 인수2))
my_sum = sum_num(3,5)
print(my_sum)
print(sum_num(1,2))
```
