# 1.람다함수

1) 람다함수 정의

```python
# 함수명 = lambda 매개변수1, 매개변수2: 반환식
f = lambda x, y: x + y
# 람다함수 호출 함수명(인수1, 인수2)
print(f(1, 2))
print(f([1,2],[3,4]))
print(f('py','thon'))

```

2) 람다함수 기본값

```python
f = lambda x, y = 3: x + y
# 인수를 1개만 설정시에는 매개변수에 설정된 기본값이 사용됨
print(f(1))
# 인수를 2개 설정시에는 기본값이 인수로 바뀌어 계산됨
print(f(2, 4))
```

5) 최대, 최소

```python
list1 = [[9,1], [8,2], [7,3], [6,4]]
max1 = max(list1, key = lambda x: x[0]) # lambda x: x[0] 첫번째 요소의 값으로 검색을 시작함
max2 = max(list1, key = lambda x: x[1]) # lambda x: x[1] 두번째 요소의 값으로 검색을 시작함

print(max1)
print(max2)

min1 = min(list1, key = lambda x: x[0])
min2 = min(list1, key = lambda x: x[1])
print(min1)
print(min2)
```

6) 정렬

```python
list2 = ['가','각','감','값','+','!','/','간','강',']']
list2.sort(key = lambda x: x) #오름차순 정렬
print(list2)
# 파이썬에서 사용되는 모든 문자열은 유니코드
# 문자열을 정렬할 때 각 문자들을 유니코드 값으로 변경하고 인스턴스 끼리 비교를 하게됨

print([ord(i) for i in list2])
```

7) 복합정렬

```python
l = [[3,'d'], [5, 'c'], [1, 'a'], [0,'b'], [0, 'a']]
l1 = sorted(l, key = lambda x: -x[0])
l2 = sorted(l, key = lambda x: x[1])
l3 = sorted(l, key = lambda x: [x[0], x[1]])
# key = lambda x: x 사용하면 첫번째 요소를 기준으로 정렬하게됨
# 특정요소를 기준으로 정렬한다면 어떤값을 기준으로 정렬할 지 표기를 해주면됨
print(l)
print()
print(l1)
print()
print(l2)
print(l3)
```

# 2. 맵 함수

# 4. 변수

1) 지역변수와 전역변수

```python
# 함수 밖에서 value에 100을 대입
value = 100 # 전역변수
def changeValue():
    # 함수 안에서 value 값을 변경 20
    value = 20 # 지역변수
    print(f'value = {value}')
changeValue()
print(f"value = {value}") # 전역변수의 값은 변경되지 않음
```

```python
# 전역변수
value = 100
def changeValue():
    global value
    # 함수 내에서 전역변수의 value 값을 변경
    value = 20

changeValue()
print(f"value = {value}")
```

# 5. 모듈

1) random 모듈

```python
import random
print(random.random()) # 0이상 1미만의 숫자 중에서 아무숫자나 하나를 뽑아서 반환하는 역할(돌려주는 역할)
print(random.randrange(1,7)) # 1부터 6까지 숫자중에서 아무 숫자나 하나 뽑아서 반환하는 역할
 
alph = ['a','b','c','d','e']
print(alph)
random.shuffle(alph) # .shuffle()시퀀스를 뒤죽박죽으로 섞어놓은 함수
print(alph)

print(random.choice(alph))

print(random.random()) # 0.0 <= x < 1.0 실수형
print(random.uniform(3.5, 7.9)) # a <= x < b 실수형
print(random.randrange(10)) # 0 <= x < a 정수형
print(random.randrange(3, 9)) # a <= < x < b 정수형
print(random.randint(5, 10))
```

```python
# 로또 추첨기 만들기
# 1부터 45 수중에 랜덤한 6개 숫자를 뽑아 출력하기
import random
list1 = []
num = random.randint(1, 45)
for i in range(6):
    while num in list1:
        num = random.randint(1, 45)
    list1.append(num)

print(list1)



```
