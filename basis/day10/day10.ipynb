```python
class Person: # Person 이라는 클래스 정의
    def getName(self): # 사람의 이름을 정의하는 메서드 정의
        return self.name # 사람의 이름을 나타내는 name 데이터 속성을 정의

    def getAge(self): # 사람의 나이를 정의하는 메서드 정의
        return self.age # 사람의 나이를 나타내는 age 데이터 속성을 정의

# 속성 이용
pr = Person() # 인스턴스 작성
# 데이터 속성에 값을 대입
pr.name = "김양진"
pr.age = 22
# 메서드 호출
print(pr.getName()) # 객체.첫 번째 메서드명
print(pr.getAge())  # 객체.두 번째 메서드명
```

```python
class Person:
    eye = 2
    nose = 1
    mouse = 1
    name = ''
    age = 0
    def __init__(self, name, age): # 인스턴스가 작성될 때 처음에 반드시 처리되는 메서드(생성)
    # 첫 번째 매개변수는 self, 자기자신으로 들어간다
        self.name = name # Person클래스의 name이라는 속성, init()함수로부터 받은 name으로 할당하겠다
        self.age = age  # self.name, self.age이 속성

    #사람의 이름을 정의하는 메서드 getName
    def getName(self):
        print(f"나는 {self.name}입니다.")
    # 사람의 나이를 정의하는 메서드 getAge
    def getAge(self):
        print(f"제 나이는 {self.age}세입니다.")

# 인스턴스 작성
pr1 = Person("김양진", 22) # 객체를 만들면서 바로 데이터 속성값을 넣어줄 수 있음 : 생성자를 만들었을 때
pr2 = Person("짱구", 5)

# 메서드 호출
pr1.getName()
pr1.getAge()
pr2.getName()
pr2.getAge()

print(f"눈: {Person.eye}, 코: {Person.nose}, 입: {Person.mouse}")
```

```python
# Person 클래스명으로 생성자 메소드, who_am_i메소드명
class Person:
    def __init__(self, name):
        self.name = name
        print(f"이름: {self.name}")

    def who_am_i(self, age, address, phone):
        self.age = age
        self.address = address
        self.phone = phone
        print(f"나이: {self.age}, 주소: {self.address}, 전화번호: {self.phone}")

pr1 = Person("김양진")
pr1.who_am_i(22, "서울","010-1234-5678")


```

> 클래스 메서드와 정적메서드

```python
# 클래스 메서드 : 클래스 변수를 사용하는 메서드
# 1. 인스턴스 또는 클래스로 호출함
# 2. 생성된 인스턴스가 없어도 호출 가능
# 3. @classmethod 데코레이터를 표시하고 작성
# 4. 매개변수 self를 사용하지 않고 cls를 사용
class Sample:
    count = 0

    def __init__(self):
        Sample.count += 1

    @classmethod
    def counting(cls):
        print(f"{cls.count}개의 객체가 생성되었습니다.")

s1 = Sample()
print(s1.count, Sample.count)
Sample.counting()
s2 = Sample()
print(s2.count, Sample.count)
Sample.counting()
s3 = Sample()
print(s3.count, Sample.count)
Sample.counting()
print("최종: ",s1.count,s2.count,s3.count)
Sample.counting()
```
