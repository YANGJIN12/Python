```python
#리스트 > 인덱싱, 슬라이싱 가능, 시퀀스자료형
list1 = [1,2,3,4,5]
print(list1, type(list1))
print(list1[1], list1[-4])
print(list1[1:4])

list2 = ['가','나','다']
print(list2[2])

list3 = [5,6,7,8,9]
print(list1 + list3)
print(list3 * 3)

list4 = [1.1, True, '가', 3.3, '나', 6.5, 'python']
print(list4 )
print(True in list4)
print(False in list4)
print(False not in list4)
```

1.조건문 if문, if~else문

0) 조건문을 위한 논리연산

```python
#and 연산 > 두 값이 모두 True일 때만 True를 반환
print(True and True)
print(True and False)
print(False and True)
print(False and False)

num1 = 40
print((num1 > 30) and (num1 < 50))
```

```python
#or 연산 > 두 값중 하나가 True이면 True를 반환  
print(True or True)
print(True or False)
print(False or True)
print(False or False)

age = 25
print((age < 15) or (age > 30))

num2 = 30
print((num2 % 10 == 0) or (num2 % 4 == 0))
```

```python
#not 연산 > True이면 False 반환, False이면 True 반환
print(not True)
print(not False)
print(not(30 > 5))
```

1) if문

```python
x = 7
if x == 10: # if 옆에 조건식이 참일때만
    print("10입니다.") # 들여쓰기된 부분이 실행됨
print("10이 아닙니다.") #들여쓰기되지 않는 부분이 실행됨
```

```python
today = '월요일'
if today == '토요일':   # SyntaxError: 문법에러 > 콜론 미사용
    print("게임을 한다") # IndentationError : 들여쓰기 에러 > if문 다음문장은 무조건 들여쓰기를 해야함
print("공부 시작")
```

```python
num1 = -1
if num1 > 3: # 첫번째 if문 조건이 참일 때만 아래로 내려가서 문장을 실행한다
    print("num1은 3보다 큽니다.")
if num1 < 0: # 들여쓰기 위치에 따라 결과값이 달라질 수 있음
    print("num1은 0보다 작습니다!")
```

```python
# 사용자가 입력한 수가 100이면 100이야를 출력, 1000이면 1000이야를 출력
# if문만 사용해서 출력

num = int(input("숫자 입력: "))
if num == 100:
    print("100이야")
if num == 1000:
    print("1000이야")
```

2) if~else문

```python
x = int(input("숫자 입력: "))
if x == 10: # if문의 조건식이 참일 때
    print("10입니다.") # 출력됨
else: # if문 옆에 조건식이 거짓일 때 내려옴
    print("10이 아닙니다.") # if문의 조건이 거짓일 때 출력됨

x = int(input("숫자 입력: "))
if x == 5:
    print("5입니다.") 
else: 
    print("5가 아닙니다.")


if x > 3 and x < 7:
    print("3보다 크고 7보다 작습니다.")
else:
    print("3이하 7이상입니다.")

int1 = int(input("정수 입력: "))
if (int1 % 2 == 0 or int1 % 3 == 0):
    print("2와 3으로 나누어 떨어집니다.")
else:
    print("2와 3으로 나누어떨어지지 않습니다.")
```
