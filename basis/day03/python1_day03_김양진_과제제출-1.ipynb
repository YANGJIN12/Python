1. 사다리꼴의 면적을 구해서 설명을 화면에 같이 표시해주는 프로그램 만들기

```python
#윗변이 2cm, 아랫변이 3cm, 높이가 4cm인 사다리꼴의 면적을 출력하기
upperbase = 2
baseline = 3 
high = 4 
#사다리꼴의 면적 구하는 공식 = (윗변 + 아랫변) x 높이 / 2
trapezoid = (upperbase + baseline) * height / 2
print('사다리꼴의 면적 구하는 공식 = (윗변: {}cm+ 아랫변: {}cm) x 높이: {} / 2 = {}'.format(upperbase, baseline, high, trapezoid))
print("사다리꼴 면적은",trapezoid,"cm²")
```

2. 아르바이트를 할 때의 수당은 일반적으로 일한 시간에 따라 계산하는 "시급"으로 받습니다.
시급과 노동시간 두가지를 입력받아서 수당을 계산하는 프로그램 만들기
(시급과 노동시간을 입력할 수 있는 변수 만들고 두 변수는 사용자로부터 입력받도록 할것)

```python
#시급은 얼마입니까? 8000
#몇시간 일했습니까? 35
#시급 8000원으로 35시간 일했으므로 수당은 280000원입니다.
hourlypay  = int(input("시급: "))
workinghours = int(input("노동시간: "))
benefit = hourlypay * workinghours
print("시급 {}원으로 {}시간 일했으므로 수당은 {}원입니다.".format(hourlypay, workinghours, benefit))
print(f"시급 {hourlypay}원으로 {workinghours}시간 일했으므로 수당은 {benefit}원입니다.")
```
